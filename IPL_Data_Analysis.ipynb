```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://docs.google.com/spreadsheets/d/1eUrPYL32VykYfRWIspYoBlWJk2g2CRM1e9nTX351_ds/export?format=csv&gid=1039833698"
df = pd.read_csv(url)
print(df.head())
```

```python
df.info()
df.describe(include='all') #Gives General idea to identify the top performer
```

```python
df.isnull().sum()
```

```python
df['isWide'] = df['isWide'].fillna(0)
df['isNoBall'] = df['isNoBall'].fillna(0)
df['Byes'] = df['Byes'].fillna(0)
df['LegByes'] = df['LegByes'].fillna(0)
df = df.drop('Penalty',axis=1)
df['dismissal_kind'] = df['dismissal_kind'].fillna('Not Out')
df['player_dismissed'] = df['player_dismissed'].fillna('Not Dismissed')
```

```python
df.isnull().sum()
```

```python
batsman_total = df.groupby('batsman')['batsman_runs'].sum()
print(batsman_total)
df['runs'] = df['batsman'].map(batsman_total)
display(df[['runs','batsman']])
```

```python
#Rank 1 is given most weightage in this Ranking and is based on Total Runs Scored by a batsman in IPL

df_top_batsman = df.sort_values (by = 'runs', ascending = False)
df_top_batsman = df_top_batsman.drop_duplicates(subset='batsman')
df_top_batsman = df_top_batsman[['batsman','runs']]
df_top_batsman['rank1'] = range(1, len(df_top_batsman) + 1)
df_top_batsman = df_top_batsman[df_top_batsman['runs']>100]
df_top_batsman
```

```python
#Graph of Runs vs Rank 1

plt.scatter(df_top_batsman['rank1'], df_top_batsman['runs'])
plt.xlabel('Rank 1')
plt.ylabel('Runs')
plt.title('Scatter Plot of Runs vs Rank 1')
plt.show()
```

```python
#Rank 2 is based on the Strike Rate of the Batsmen in IPL, This rank is given lesser weightage than Rank 1

batsman_counts = df['batsman'].value_counts()
df_top_batsman['balls'] = df['batsman'].map(batsman_counts)
df_top_batsman = df_top_batsman.sort_values(by='balls', ascending=False)
df_top_batsman = df_top_batsman[df_top_batsman['balls']>100]
df_top_batsman = df_top_batsman.drop_duplicates(subset='batsman')


df_top_batsman['strike_rate'] = (df_top_batsman['runs'] / df_top_batsman['balls']) * 100
df_top_batsman = df_top_batsman.sort_values(by='strike_rate', ascending=False)
df_top_batsman['rank2'] = range(1, len(df_top_batsman) + 1)
df_top_batsman
```

```python
#Graph of Strike Rate vs Rank 2

plt.scatter(df_top_batsman['rank2'], df_top_batsman['strike_rate'])
plt.xlabel('Rank 2')
plt.ylabel('Strike Rate')
plt.title('Scatter Plot of Strike Rate vs Rank 2')
plt.show()
```

```python
#Average Rank is the sum of Rank 1 * 3 + Rank 2 (i.e. they are in the ratio of 3:1 in terms of weightage)

df_top_batsman['avg_rank'] = df_top_batsman['rank1'] * 3 + df_top_batsman['rank2']
df_top_batsman = df_top_batsman.sort_values(by='avg_rank')
df_top_batsman['final_rank'] = range(1, len(df_top_batsman) + 1)
df_top_batsman.head(20)
```

```python
#List of Top 20 Batsman in IPL

df_top_batsman = df_top_batsman.reset_index(drop = True)
df_top_batsman[['final_rank','batsman']].head(20)
```

```python
print(df['bowler'].value_counts())
```

```python
df['out'] = df.dismissal_kind.apply(lambda x: 1 if x != 'Not Out' else 0)
#df[['out','dismissal_kind']]
#print(df[df['out']==1])
```

```python
#Rank 1 for Bowlers is Based on the Number of Wickets that they have taken

bowler_wickets = df.groupby('bowler')['out'].sum()

df['wickets'] = df['bowler'].map(bowler_wickets)

df_top_bowler = df.sort_values(by='wickets', ascending=False)
df_top_bowler = df_top_bowler.drop_duplicates(subset='bowler')
df_top_bowler = df_top_bowler[['bowler','wickets']]

df_top_bowler = df_top_bowler.drop_duplicates(subset='bowler')
df_top_bowler = df_top_bowler[df_top_bowler['wickets']>10]
df_top_bowler['rank1'] = range(1, len(df_top_bowler) + 1)
df_top_bowler
```

```python
#Graph of Wickets vs Rank 1 for Bowlers

plt.scatter(df_top_bowler['rank1'], df_top_bowler['wickets'])
plt.xlabel('Rank 1')
plt.ylabel('Wickets')
plt.title('Scatter Plot of Wickets vs Rank 1')
plt.show()
```

```python
bowls = df['bowler'].value_counts()
df_top_bowler['balls'] = df['bowler'].map(bowls)

runs_taken = df.groupby('bowler')['batsman_runs'].sum()
df_top_bowler['runs_conceded'] = df['bowler'].map(runs_taken)
df_top_bowler['economy'] = df_top_bowler['runs_conceded'] / (df_top_bowler['balls'] / 6)
df_top_bowler = df_top_bowler.sort_values(by='economy')
df_top_bowler['rank2'] = range(1,len(df_top_bowler)+1)
df_top_bowler
```

```python
#Graph of Economy vs Rank 2 for Bowlers

plt.scatter(df_top_bowler['rank2'], df_top_bowler['economy'])
plt.xlabel('Rank 2')
plt.ylabel('Economy')
plt.title('Scatter Plot of Economy vs Rank 2')
plt.show()
```

```python
#Average Rank is the sum of Rank 1 * 5 + Rank 2 * 2 (i.e. they are in the ratio of 5:2 in terms of weightage)

df_top_bowler['avg_rank'] = (df_top_bowler['rank1'] * 5) + (df_top_bowler['rank2'] *2)
df_top_bowler = df_top_bowler.sort_values(by='avg_rank')
df_top_bowler['final_rank'] = range(1, len(df_top_bowler) + 1)
df_top_bowler
```

```python
#List of Top 20 Batsman in IPL

df_top_bowler = df_top_bowler.reset_index(drop = True)
df_top_bowler[['final_rank','bowler']].head(20)
```

```python
#Top 10 Batsmen in Powerplay (Sorted Based on Runs as that shows a Powerplay Specialist who can give their team a Headstart)

df_power = df[df['over_ball']<=5.0]

df_power_bat = df_power.groupby('batsman')['batsman_runs'].sum()
df_power_bat = df_power_bat.sort_values(ascending=False)
print(df_power_bat.head(10))
df_power.head()
```

```python
#Top 10 Bowlers in Powerplay (Sorted Based on Wicket as that's a better way to evaluate than Economy in Powerplay)

df_power_bowl = df_power.groupby('bowler')['out'].sum()
df_power_bowl = df_power_bowl.sort_values(ascending=False)
df_power_bowl.head(10)
```

```python
#Top 10 Batsmen in Death Overs (Sorted Based on Strike as that shows a Finisher who can improve the team's Run Rate)

df_death = df[df['over_ball']>=16.0]

balls_faced = df_death['batsman'].value_counts()

runs_scored = df_death.groupby('batsman')['batsman_runs'].sum()

df_death['death_runs'] = df_death['batsman'].map(runs_scored)
df_death['ball'] = df_death['batsman'].map(balls_faced)


df_death['strike_rate'] = (df_death['death_runs'] / df_death['ball']) * 100


df_death_bat = df_death.groupby('batsman')['strike_rate']
df_death_bat = df_death[['batsman','strike_rate','ball']]
df_death_bat = df_death_bat[df_death_bat['ball']>60]
df_death_bat = df_death_bat[['batsman','strike_rate']]
df_death_bat = df_death_bat.drop_duplicates(subset='batsman')

df_death_bat = df_death_bat.sort_values(by='strike_rate', ascending=False)

df_death_bat = df_death_bat.reset_index(drop=True)

df_death_bat.head(10)
#df_death
```

```python
bowls = df_death['bowler'].value_counts()
#print(df_death['bowler'].value_counts())
df_death['balls'] = df_death['bowler'].map(bowls)
runs_conceded = df_death.groupby('bowler')['batsman_runs'].sum()

df_death['death_runs_conceded'] = df_death['bowler'].map(runs_conceded)
#df_death['balls_thrown'] = df_death['bowler'].map(balls_delivered)

df_death = df_death.drop_duplicates(subset='bowler')
df_death = df_death[df_death['balls']>60]
df_death['economy'] = (df_death['death_runs_conceded'] / df_death['balls'])*6
df_death[['bowler','economy','balls']]
```

```python
#Top 10 Bowlers in Death Overs (Sorted Based on Economy as that shows a Bowler who concedes less runs even in death overs)

df_death_bowl = df_death.groupby('bowler')['economy']
df_death_bowl = df_death[['bowler','economy']]

df_death_bowl = df_death_bowl.drop_duplicates(subset='bowler')
df_death_bowl = df_death_bowl.sort_values(by='economy', ascending=True)

df_death_bowl = df_death_bowl.reset_index(drop=True)

df_death_bowl.head(10)
```

```python
#Action Insight 1: Top Match-ups (Based on Max Balls Faced)

matchups = df.groupby(['batsman','bowler']).size().reset_index(name='balls_faced')

runs = df.groupby(['batsman','bowler'])['batsman_runs'].sum().reset_index(name='runs_scored')
matchups = matchups.merge(runs, on=['batsman','bowler'])

out = df.groupby(['batsman','bowler'])['out'].sum().reset_index(name='wickets_taken')
matchups = matchups.merge(out, on=['batsman','bowler'])

top_matchups = matchups.sort_values(by='balls_faced', ascending=False).reset_index(drop=True)

top_matchups.head(10)
```

```python
#Visual Plots

plt.scatter(matchups['balls_faced'], matchups['runs_scored'])
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.title('Plot of Runs Scored vs Balls Faced')
plt.show()

plt.scatter(matchups['wickets_taken'], matchups['runs_scored'])
plt.xlabel('Balls Faced')
plt.ylabel('Wickets Scored')
plt.title('Plot of Wickets vs Balls Faced')
plt.show()
```

```python
#Action Insight 2: Max Wickets in First Over

df_first = df[df['over_ball']<=1.0]
df_first_wickets = df_first.groupby('bowler')['out'].sum()
df_first_wickets = df_first_wickets.sort_values(ascending=False)

df_first_wickets.head(3)
```

```python
#Graph Plot

plt.scatter(df_first_wickets.index, df_first_wickets.values)
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.title('Plot of Wickets vs Bowler')
plt.show()

#This shows that after the first 2 entries their is a steep fall indicating that not many bowlers specialize in first over wickets
```

```python
#Action Insight 3: Maximum Six hitter's Most favourable mode of dismissal

df_six = df[df['batsman_runs']==6]
df_six_bat = df_six.groupby(['batsman']).size().reset_index(name='six_count')
df_six_bat = df_six_bat.sort_values(by='six_count', ascending=False)

df_six_bat = df_six_bat.reset_index(drop=True)
df_six_bat.head(10)
```

```python
top_six = df_six_bat.head(10)
dismissals = df[df['player_dismissed'].notna() & df['batsman'].isin(top_six['batsman'])].groupby(['batsman','dismissal_kind']).size().reset_index(name='count')

for type in dismissals['dismissal_kind'].unique():
    dismissals_type = dismissals[dismissals['dismissal_kind']==type]
    counts = dismissals_type.set_index('batsman')['count']   # lookup
    top_six[type] = top_six['batsman'].map(counts).fillna(0).astype(int)

top_six = top_six.drop(columns = 'Not Out')
top_six
```

```python
dismissal_cols = [col for col in top_six.columns if col not in ['batsman','six_count']]

dismissal_sums = top_six[dismissal_cols].sum()

# Plot
plt.figure()
dismissal_sums.plot(kind='bar')
plt.ylabel("Total Dismissals")
plt.title("Total Dismissals by Type (Top 10 Six Hitters)")
plt.show()

#This gives the insight that top six hitters and most frequently caught
```
